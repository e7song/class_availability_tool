##### Demonstration of Idea for COGS

```python
import requests
from bs4 import BeautifulSoup
```

```python
all_class = 'https://catalog.ucsd.edu/courses/COGS.html'
tentative_offerings = 'https://cogsci.ucsd.edu/undergraduates/courses/index.html#COGS-Course-Offerings-2024-25'
```

```python
def get_url(url):
    r = requests.get(url)
    if r.status_code == 200: # success
        soup = BeautifulSoup(r.text, 'html.parser')
        return soup
    else:
        return f'Status Code: {r.status_code}'
```

```python
soup = get_url(all_class)
```

```python
soup.find_all('h2')[1].text
```

```python
courses_dct = {}
for h2_tag in soup.find_all('h2'):
    subheading = h2_tag.get_text()
    courses_dct[subheading] = h2_tag.get_next_siblings()
```
